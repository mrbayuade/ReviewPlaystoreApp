<b>Scrape Google Play Reviews in 4 simple steps using Python<b>

Step 1 : Import required packages

```python
from google_play_scraper import app
from google_play_scraper import Sort, reviews_all
import pandas as pd
import numpy as np
```

Step 2 : Find the App Id in Google Play Store
- I this project I use Indodax app (url : <a>https://play.google.com/store/apps/details?id=id.co.bitcoin)</a>.
- You can see the App Id in the URL after 'id=' for Indodax.
- You can download reviews in different languages using ISO-Code (<a>https://en.wikipedia.org/wiki/List_of_ISO_639-1_codes</a>) for language.
  Example, lang = 'fr' can be used to download for French reviews.


```python
df_reviews = reviews_all(
    'id.co.bitcoin', # find in the url after '='
    sleep_milliseconds = 0, # defaults to 0
    lang = 'id', # indonesian language
    country = 'id', # indonesia
    sort = Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

Step 3: Put the Reviews into Pandas DataFrame

```python
df = pd.DataFrame(np.array(df_reviews),columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))
df.head()
```

Step 4: Find Value Counts for Star 1, 2, 3, 4, and 5

```python
df.value_counts('score')
```

You can store review dataset to excel by following code :

```python
df.to_excel('ReviewIndodax.xlsx',index = False)
```
